```python
#Import the relevant libraries
import pandas as pd
import numpy as np
import plotly.plotly as py
import plotly.graph_objs as go
import cufflinks as cf
from plotly.offline import init_notebook_mode,iplot,plot
```

```python
#http://databank.worldbank.org/data/reports.aspx?source=world-development-indicators
#Import the dataset
countries = pd.read_excel('https://s3-eu-west-1.amazonaws.com/4thyearproject/4thYearProject/ProjectDataFiles/Data_Extract_From_World_Development_Indicators_GDP.xlsx')
```

```python
#Show data set
countries
```

```python
#Rename the column headings
countries.rename(columns = {'Country Code':'Country_Code','1990 [YR1990]':'1990','2000 [YR2000]':'2000','2001 [YR2001]':'2001','2002 [YR2002]':'2002','2003 [YR2003]':'2003','2004 [YR2004]':'2004','2005 [YR2005]':'2005','2006 [YR2006]':'2006','2007 [YR2007]':'2007','2008 [YR2008]':'2008','2009 [YR2009]':'2009','2010 [YR2010]':'2010','2011 [YR2011]':'2011','2012 [YR2012]':'2012','2013 [YR2013]':'2013','2014 [YR2014]':'2014','2015 [YR2015]':'2015', '2016 [YR2016]':'2016'}, inplace=True) 
```

```python
#selecting the coumns I want to keep and drop the rest
```

```python
#pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
countries
```

```python
#making a copy of the new data set
```

```python
countries = countries[keep].copy()
```

```python
countries
```

```python
#removing any cells that containe invalid data and using inplace to make to change perminent
```

```python
countries.dropna(how='any', inplace=True)
```

```python
#setting the country code colmn as the first index
```

```python
countries.set_index('Country_Code', inplace=True)
```

```python
countries=countries.T
```

```python
#Atempting to change the scientific notations

#countries.USA = countries.USA.astype(np.int64)
#countries.GBR = countries.GBR.astype(np.int64)
#countries.CHN = countries.CHN.astype(np.int64)
#countries.IRL = countries.IRL.astype(np.int64)
#countries.RUS = countries.RUS.astype(np.int64)
#df.astype(np.int64, copy=True) #inplace=True 
countries['USA'] = countries.USA.map(lambda x: '{:.0f}'.format(x))
countries['GBR'] = countries.GBR.map(lambda x: '{:.0f}'.format(x))
countries['CHN'] = countries.CHN.map(lambda x: '{:.0f}'.format(x))
countries['IRL'] = countries.IRL.map(lambda x: '{:.0f}'.format(x))
countries['RUS'] = countries.RUS.map(lambda x: '{:.0f}'.format(x))
```

```python
#Viewing the changes
countries
```

```python
#Plotting the results iplot allows for the graph to appear in jupyter noebook
#kind indicate the type of graph
#xTitle places a label for the x axis as does yTitle places a label for the y axis
#title is the main title at the top of the graph
countries.iplot(kind='line',xTitle='Year',yTitle='GDP',title='GDP Between 2000 and 2015')
```

```python
#https://www.bea.gov/itable/iTable.cfm?ReqID=70&step=1#reqid=70&step=1&isuri=1
#Reading in the data from the csv file stored in S3
df3 = pd.read_csv('https://s3-eu-west-1.amazonaws.com/4thyearproject/4thYearProject/ProjectDataFiles/gdp_by_state.csv', skiprows=4, header=None)
```

```python
#Viewing the data frame df3
df3.head()
```

```python
#Viewing the results
df3.head()
```

```python
df3.head()
```

```python
#Viewing the changes made
df3
```

```python
#Resetting the index
df3 = df3.reset_index(drop=True)
```

```python
df3.head()
```

```python
df3.head()
```

```python
#Checking the changes made so far
df3
```
